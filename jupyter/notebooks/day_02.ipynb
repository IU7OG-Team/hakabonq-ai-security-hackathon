```python
import pandas as pd
import numpy as np
import random as rnd
import math

train_df = pd.read_csv('train_longevity.csv')
test_df = pd.read_csv('test_longevity.csv')
```

```python
# Задание №1

avg_train = train_df.dropna(subset=['Age'])['Age'].astype(int).mean()
avg_test = test_df.dropna(subset=['Age'])['Age'].astype(int).mean()

train_df['Age'] = train_df['Age'].map(lambda x: avg_train if math.isnan(x) else x)
test_df['Age'] = test_df['Age'].map(lambda x: avg_test if math.isnan(x) else x)
train_df
```

```python
# Задание №2

train_df = train_df.drop(['MedExam'], axis='columns')
test_df = test_df.drop(['MedExam'], axis='columns')
test_df
```

```python
# Задание №3

avg_train = train_df['Activity'].mean()
avg_test = test_df['Activity'].mean()

def sport_predictor(sport, activity, education, avg):
    if type(sport) == float:
        if activity > avg and education == 1:
            return '+'
        else:
            return '-'
            
    return sport

train_df['Sport'] = train_df.apply(lambda x: sport_predictor(x['Sport'], x['Activity'], x['Education'], avg_train), axis=1)
test_df['Sport'] = test_df.apply(lambda x: sport_predictor(x['Sport'], x['Activity'], x['Education'], avg_test), axis=1)
test_df
```

```python
# Задание №4

train_df = train_df.drop(['Region'], axis='columns')
test_df = test_df.drop(['Region'], axis='columns')
test_df
```

```python
# Задание №5

train_df = train_df.drop(['Id'], axis='columns')
train_df
```

```python
# Задание №6

train_df['Family'] = train_df.apply(lambda x: x['Children'] + x['Pet'], axis=1)
test_df['Family'] = test_df.apply(lambda x: x['Children'] + x['Pet'], axis=1)
train_df
```

```python
# Задание №7

train_df['IsAlone'] = train_df.apply(lambda x: not x['Pet'] and not x['Children'], axis=1)
test_df['IsAlone'] = test_df.apply(lambda x: not x['Pet'] and not x['Children'], axis=1)
train_df
```

```python
# Задание №8

def create_sign(age, education):
    if education == 3 and age < 75:
        return 'MiddleAged_Poor'
    elif education == 3 and age >= 75:
        return 'Old_Poor'
    elif education == 2 and age < 75:
        return 'MiddleAged_MiddleIncome'
    elif education == 2 and age >= 75:
        return 'Old_MiddleIncome'
    elif education == 1 and age < 75:
        return 'MiddleAged_Rich'
    else:
        return 'Old_Rich'

train_df['SocialStatus'] = train_df.apply(lambda x: create_sign(x['Age'], x['Education']), axis=1)
test_df['SocialStatus'] = test_df.apply(lambda x: create_sign(x['Age'], x['Education']), axis=1)
train_df
```

```python
# Задание №9

def create_sign(age):
    if age < 70:
        return 1
    elif age <= 72:
        return 2
    elif age <= 74:
        return 3
    elif age <= 76:
        return 4
    elif age <= 78:
        return 5
    else:
        return 6
    
train_df['AgeInterval'] = train_df.apply(lambda x: create_sign(x['Age']), axis=1)
test_df['AgeInterval'] = test_df.apply(lambda x: create_sign(x['Age']), axis=1)
train_df
```
